**Mudando valores linha por linha - iterrows()**

```python
cabine = pd.read_csv('titanic/train.csv')
```

```python
cabine['Cabin']
```

```python
for idx, _ in cabine[['Cabin']].dropna().iterrows():
    cabine_individual = cabine['Cabin'].at[idx]
    cabine['Cabin'].at[idx] = cabine_individual[0]
```
