### Módulo Estatística

**Estatística**

"A estatística é o campo da matemática que relaciona fatos e números em que há um conjunto de métodos que nos possibilita coletar dados e analisá-los, assim sendo possível realizar alguma interpretação deles. A estatística é dividida em duas partes: descritiva e inferencial. A estatística descritiva é caracterizada pela organização, análise e apresentação dos dados, enquanto a estatística inferencial tem como característica o estudo de uma amostra de determinada população e, com base nela, a realização de análises e a apresentação de dados."

Fonte: https://brasilescola.uol.com.br/matematica/estatistica-2.htm

**Histograma**

"O histograma, também conhecido como distribuição de frequências, é a representação gráfica em colunas ou em barras de um conjunto de dados previamente tabulado e dividido em classes uniformes ou não uniformes. A base de cada retângulo representa uma classe."

Fonte: https://www.google.com/search?client=ubuntu&channel=fs&q=histograma&ie=utf-8&oe=utf-8

**Importação das bibliotecas**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

A biblioteca ***pandas*** é utilizada para manipulação dos dados, bem como frames. Com relação à biblioteca ***matplotlib***, esta é utilizada para a visualização dos dados.

Neste exemplo iremos trabalhar com a relação entre distâncias de casa para o trabalho, em termos cotidianos.

```python
km = pd.Series([4,6,6,7,11,13,27,35,36,36,42,43,45,49,18,18,21,24,26])
```

```python
lista = km.values
```

```python
def calcula_ticks(lista, barras):
    lista.sort()
    maior_valor = lista[-1]
    menor_valor = lista[0]
    
    intervalo = (maior_valor - menor_valor)/barras
    ticks = [menor_valor]
    ultimo = menor_valor+intervalo
    ticks.append(ultimo)
    
    for x in range(barras-1):
        ultimo += intervalo
        ticks.append(ultimo)
    return ticks
```

```python
barras = 6
km.plot.hist(bins=barras, rwidth=0.95)
ticks = calcula_ticks(lista, barras)
plt.xticks(ticks)
plt.title('Distância de Casa para Trabalho')
plt.xlabel('km')
plt.ylabel('Frequência')
plt.grid(axis='y')
plt.show()
```

**Frequência Relativa**

"Em estatística denomina-se frequência relativa o resultado obtido da divisão entre a frequência absoluta - o valor que é observado na população - e a quantidade de elementos da amostra. Geralmente é apresentada na forma de porcentagem, a partir da multiplicação por 100."

Fonte: https://www.google.com/search?client=ubuntu&channel=fs&q=Frequencia+relativa&ie=utf-8&oe=utf-8

```python
len(km.values)
```

```python
def frequencia_relativa(total, yticks):
    freq_rel = []
    for item in yticks:
        x = item/total
        freq_rel.append("{0:.2f}%".format(x*100))
    return freq_rel
```

```python
def gerar_histograma(barras,data_series,titulo,unidade):
    lista = data_series.values
    ax = data_series.plot.hist(bins=barras, rwidth=0.95)

    yticks = ax.get_yticks()
    total = len(lista)
    freq_rel = frequencia_relativa(total,yticks)
    ax.set_yticklabels(freq_rel)
    
    fig = plt.gcf()
    fig.set_size_inches(10,5)
    fig.set_dpi(150)
    
    ticks = calcula_ticks(lista, barras)
    plt.xticks(ticks)
    plt.title(titulo)
    plt.xlabel('cm')
    plt.ylabel('Frequência')
    plt.grid(axis='y')
    plt.show()
```

```python
gerar_histograma(5,km,'Distância entre Casa e Trabalho','km')
```

**Coeficiente de variação**

"Em teoria das probabilidades e estatística, o coeficiente de variação, também conhecido como desvio padrão relativo, é uma medida padronizada de dispersão de uma distribuição de probabilidade ou de uma distribuição de frequências."

Fonte: https://www.google.com/search?client=ubuntu&channel=fs&q=coeficiente+de+varia%C3%A7%C3%A3o&ie=utf-8&oe=utf-8

```python
cv_altura_mulheres = dp_altura_mulheres/media_altura_mulheres
```

```python
cv_altura_mulheres
```

```python
def pct(x):
    return "{0:.2f}%".format(x*100)
```

```python
pct(cv_altura_mulheres)
```

```python
cv_altura_homens = dp_altura_homens/media_altura_homens
```

**Covariância e Coeficiente de Correlação***

"Em teoria da probabilidade e na estatística, a covariância, ou variância conjunta, é uma medida do grau de interdependência numérica entre duas variáveis aleatórias. Assim, variáveis independentes têm covariância zero."

"Em estatística descritiva, o coeficiente de correlação de Pearson, também chamado de "coeficiente de correlação produto-momento" ou simplesmente de "ρ de Pearson" mede o grau da correlação entre duas variáveis de escala métrica. Este coeficiente, normalmente representado por ρ assume apenas valores entre -1 e 1."

```python
top_cinco = data.head()
top_cinco
```

```python
top_cinco = top_cinco[['Altura cm', 'Peso kg']]
```

```python
media_altura = top_cinco['Altura cm'].mean()
media_altura
```

```python
media_peso = top_cinco['Peso kg'].mean()
media_peso
```

```python
top_cinco_lista = []
```

```python
for i in range(0, len(top_cinco)):
    top_cinco_lista.append((top_cinco['Altura cm'][i]-media_altura)*(top_cinco['Peso kg'][i]-media_peso))
```

```python
top_cinco_lista
```

```python
top_cinco_cov = sum(top_cinco_lista)/(len(top_cinco)-1)
top_cinco_cov
```
